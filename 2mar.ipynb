## Ans 3)

```python
# The subplot() function in Matplotlib is used to create multiple plots in the same figure, allowing you to arrange plots in a grid-like structure
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='orange')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot 4')
plt.legend()

plt.tight_layout() 
plt.show()
```

## Ans 4)

```python
# A bar plot, also known as a bar chart, is a graphical representation of data using rectangular bars to show the relationship between different categories and their corresponding values. 
# Each bar in the plot represents a category, and the height of the bar represents the value associated with that category. 
# Comparison of Categories
# Categorical Data Visualization
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(company, profit)
```

```python
plt.barh(company, profit)
```

## Ans 5)

```python
#A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. 
# It is used to show the spread, central tendency, and potential outliers in the data. 

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box1 and Box2')
plt.grid(True)
plt.show()
```
